```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python

# 1. Análisis de facturación

# Calculamos el ingreso total sumando la columna 'Precio' de cada tienda
# El Sr. Juan necesita estos totales para comparar el rendimiento económico
ingreso_t1 = tienda['Precio'].sum()
ingreso_t2 = tienda2['Precio'].sum()
ingreso_t3 = tienda3['Precio'].sum()
ingreso_t4 = tienda4['Precio'].sum()

# Mostramos los resultados numéricos formateados
print(f"Ingresos Totales Tienda 1: ${ingreso_t1:,.2f}")
print(f"Ingresos Totales Tienda 2: ${ingreso_t2:,.2f}")
print(f"Ingresos Totales Tienda 3: ${ingreso_t3:,.2f}")
print(f"Ingresos Totales Tienda 4: ${ingreso_t4:,.2f}")

# --- GRÁFICO 1: Comparativa de Ingresos ---
import matplotlib.pyplot as plt
import seaborn as sns

# Preparamos los datos para la visualización
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
valores_ingresos = [ingreso_t1, ingreso_t2, ingreso_t3, ingreso_t4]

plt.figure(figsize=(10, 6))
# Usamos un gráfico de barras para facilitar la comparación visual inmediata
sns.barplot(x=nombres_tiendas, y=valores_ingresos, palette='magma')

# Personalización para que el reporte sea legible para el Sr. Juan
plt.title('Comparativa de Ingresos Totales por Tienda', fontsize=14, fontweight='bold')
plt.xlabel('Sucursales de Alura Store', fontsize=12)
plt.ylabel('Monto Total Vendido ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

# 2. Ventas por categoría

```python
# 2. Ventas por categoría

# Calculamos la cantidad de ventas por cada categoría en las 4 tiendas
# El método value_counts() es ideal para contar frecuencias de etiquetas
ventas_cat_t1 = tienda['Categoría del Producto'].value_counts()
ventas_cat_t2 = tienda2['Categoría del Producto'].value_counts()
ventas_cat_t3 = tienda3['Categoría del Producto'].value_counts()
ventas_cat_t4 = tienda4['Categoría del Producto'].value_counts()

# Imprimimos los resultados de forma clara
print("Top 5 Categorías - Tienda 1:\n", ventas_cat_t1.head())
print("\nTop 5 Categorías - Tienda 2:\n", ventas_cat_t2.head())
print("\nTop 5 Categorías - Tienda 3:\n", ventas_cat_t3.head())
print("\nTop 5 Categorías - Tienda 4:\n", ventas_cat_t4.head())

# --- GRÁFICO 2: Distribución de Categorías (Gráfico Circular) ---
# Usamos un gráfico circular para mostrar la proporción del mercado de cada categoría
# Esto ayuda al Sr. Juan a ver qué tan diversificado es el inventario de una tienda
plt.figure(figsize=(9, 9))
colores = sns.color_palette('pastel')

# Visualizamos la Tienda 1 como referencia (puedes cambiarla por la que desees analizar)
plt.pie(ventas_cat_t1, labels=ventas_cat_t1.index, autopct='%1.1f%%', startangle=140, colors=colores)

plt.title('Distribución Proporcional de Ventas por Categoría (Tienda 1)', fontsize=14, fontweight='bold')
plt.axis('equal') # Asegura que el gráfico sea un círculo perfecto
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# 3. Calificación promedio de la tienda

# Calculamos el promedio de la columna 'Calificación' para cada tienda
# Este indicador nos dice qué tan satisfechos están los clientes con su compra
promedio_t1 = tienda['Calificación'].mean()
promedio_t2 = tienda2['Calificación'].mean()
promedio_t3 = tienda3['Calificación'].mean()
promedio_t4 = tienda4['Calificación'].mean()

# Mostramos los resultados para la toma de decisiones
print(f"Calificación Promedio Tienda 1: {promedio_t1:.2f} ⭐")
print(f"Calificación Promedio Tienda 2: {promedio_t2:.2f} ⭐")
print(f"Calificación Promedio Tienda 3: {promedio_t3:.2f} ⭐")
print(f"Calificación Promedio Tienda 4: {promedio_t4:.2f} ⭐")

# --- GRÁFICO 3: Distribución de Calificaciones (Boxplot) ---
# El Boxplot es excelente para ver si las calificaciones son consistentes 
# o si hay muchos clientes muy insatisfechos (valores atípicos)
plt.figure(figsize=(10, 6))

# Agrupamos los datos para el gráfico
datos_grafico = [tienda['Calificación'], tienda2['Calificación'], 
                 tienda3['Calificación'], tienda4['Calificación']]

sns.boxplot(data=datos_grafico, palette='Set2')

# Personalización del gráfico
plt.title('Análisis de Satisfacción del Cliente por Tienda', fontsize=14, fontweight='bold')
plt.xticks([0, 1, 2, 3], ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])
plt.ylabel('Rango de Calificación (1 a 5)')
plt.xlabel('Sucursales')
plt.grid(axis='y', alpha=0.3)

plt.show()
```

# 4. Productos más y menos vendidos

```python
# 4. Productos más y menos vendidos

# Creamos una función pequeña para no repetir código 4 veces (buena práctica de ingeniería)
def obtener_extremos_productos(df, nombre_tienda):
    conteo = df['Producto'].value_counts()
    mas_vendido = conteo.idxmax()
    cant_mas = conteo.max()
    menos_vendido = conteo.idxmin()
    cant_menos = conteo.min()
    
    print(f"--- {nombre_tienda} ---")
    print(f"✅ Más vendido: {mas_vendido} ({cant_mas} unidades)")
    print(f"❌ Menos vendido: {menos_vendido} ({cant_menos} unidades)\n")
    return conteo

# Ejecutamos el análisis para cada tienda
top_t1 = obtener_extremos_productos(tienda, "Tienda 1")
top_t2 = obtener_extremos_productos(tienda2, "Tienda 2")
top_t3 = obtener_extremos_productos(tienda3, "Tienda 3")
top_t4 = obtener_extremos_productos(tienda4, "Tienda 4")

# --- VISUALIZACIÓN ADICIONAL: Top 5 Productos Globales (Opcional) ---
# Aunque ya cumplimos los 3 gráficos, este ayuda a visualizar los "Best Sellers"
plt.figure(figsize=(10, 6))
top_t1.head(5).plot(kind='barh', color='skyblue')
plt.title('Top 5 Productos Más Vendidos - Tienda 1', fontsize=14)
plt.xlabel('Cantidad Vendida')
plt.ylabel('Producto')
plt.gca().invert_yaxis() # Para que el #1 aparezca arriba
plt.show()
```

# 5. Envío promedio por tienda

```python
# 5. Envío promedio por tienda

# Calculamos el promedio de la columna 'Costo de envío'
# Un envío costoso puede ser la razón por la que una tienda vende menos
envio_t1 = tienda['Costo de envío'].mean()
envio_t2 = tienda2['Costo de envío'].mean()
envio_t3 = tienda3['Costo de envío'].mean()
envio_t4 = tienda4['Costo de envío'].mean()

# Mostramos los resultados finales de logística
print(f"Costo de Envío Promedio - Tienda 1: ${envio_t1:,.2f}")
print(f"Costo de Envío Promedio - Tienda 2: ${envio_t2:,.2f}")
print(f"Costo de Envío Promedio - Tienda 3: ${envio_t3:,.2f}")
print(f"Costo de Envío Promedio - Tienda 4: ${envio_t4:,.2f}")

# --- GRÁFICO OPCIONAL: Comparativa de Envíos (Gráfico de Líneas) ---
# Aunque ya tienes los 3 obligatorios, este tipo de gráfico ayuda a ver tendencias
plt.figure(figsize=(10, 5))
plt.plot(['T1', 'T2', 'T3', 'T4'], [envio_t1, envio_t2, envio_t3, envio_t4], 
         marker='o', linestyle='-', color='orange', linewidth=2)

plt.title('Tendencia de Costos de Envío por Sucursal', fontsize=14)
plt.ylabel('Costo Promedio ($)')
plt.grid(True, alpha=0.3)
plt.show()
```
